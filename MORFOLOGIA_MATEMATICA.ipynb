# CONJUNTOS

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagen en escala de grises
image = cv2.imread('CHERRY.jpg', cv2.IMREAD_GRAYSCALE)

# Verificar si la imagen se carg correctamente
if image is None:
    print("Error: La imagen no se pudo cargar.")
else:
    print("Imagen cargada correctamente.")

# Binarizacin (umbralizacin)
# Usa un umbral para convertir la imagen en binaria (0 o 255)
# cv2.THRESH_BINARY convierte pxeles por encima del umbral en 255, los dems en 0
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Definir el elemento estructurante
kernel = np.ones((10,10), np.uint8)

# Dilatacin
dilated = cv2.dilate(binary_image, kernel)

# Erosin
eroded = cv2.erode(binary_image, kernel)

# Apertura (erosin seguida de dilatacin)
opened = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel)

# Cierre (dilatacin seguida de erosin)
closed = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)

# Aumentar el tamao del plot
plt.figure(figsize=(12, 12))  # Cambia el tamao del grfico (ancho, alto)

# Mostrar resultados
titles = ['Original', 'Binarizada', 'Dilatacin', 'Erosin', 'Apertura', 'Cierre']
images = [image, binary_image, dilated, eroded, opened, closed]

for i in range(6):
    plt.subplot(3, 2, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()

```

# LATTICE

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagen en escala de grises
image = cv2.imread('MEDUSA.jpg', cv2.IMREAD_GRAYSCALE)

# Definir el elemento estructurante
kernel = np.ones((20,20), np.uint8)

# Dilatacin
dilated = cv2.dilate(image, kernel)

# Erosin
eroded = cv2.erode(image, kernel)

# Apertura (erosin seguida de dilatacin)
opened = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# Cierre (dilatacin seguida de erosin)
closed = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

# Aumentar el tamao del plot
plt.figure(figsize=(12, 12))  # Cambia el tamao del grfico (ancho, alto)
# Mostrar resultados
titles = ['Original', 'Dilatacin', 'Erosin', 'Apertura', 'Cierre']
images = [image, dilated, eroded, opened, closed]

for i in range(5):
    plt.subplot(3, 2, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```
