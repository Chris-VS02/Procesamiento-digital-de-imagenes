# BINARIZACION

```python
import cv2
import matplotlib.pyplot as plt

# Cargar imagen en escala de grises
img = cv2.imread('SEAH.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar binarizacin con umbral fijo
_, binarizada = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Mostrar resultados
plt.subplot(1,2,1), plt.imshow(img, cmap='gray'), plt.title("Original")
plt.subplot(1,2,2), plt.imshow(binarizada, cmap='gray'), plt.title("Binarizada")
plt.show()
```

```python
# Aplicar binarizacin adaptativa
binarizada_adapt = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

plt.imshow(binarizada_adapt, cmap='gray'), plt.title("Binarizacin Adaptativa")
plt.show()
```

```python
# Aplicar binarizacin de Otsu
_, binarizada_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.imshow(binarizada_otsu, cmap='gray'), plt.title("Binarizacin Otsu")
plt.show()
```

```python
img = cv2.imread('imagen.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
v = hsv[:,:,2]  # Canal de brillo

_, binarizada_v = cv2.threshold(v, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binarizada_v, cmap='gray'), plt.title("Binarizacin Canal V")
plt.show()
```
